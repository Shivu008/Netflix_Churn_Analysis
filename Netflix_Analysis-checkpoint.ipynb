```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = r'C:\Users\shiva\Downloads\projects\End to End Project\Netflix_Analytics_100k.xlsx'
df = pd.read_excel(file_path)
df.head(9)
```

```python
df.info()
df.describe()
df.columns
df.isnull().sum()
```

```python
print(df.columns.tolist())

```

```python
df.drop_duplicates(inplace=True)
df.dropna(subset=['User_Rating', 'Watch_Hours'], inplace=True)
```

```python
print("Total Users:", df['User_ID'].nunique())
print("Avg. User Rating:", round(df['User_Rating'].mean(), 2))
print("Avg. Watch Hours:", round(df['Watch_Hours'].mean(), 2))
print("Churn Rate:", round((df['Churned'] == "Yes").mean() * 100, 2), "%")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='Favorite_Genre', data=df, order=df['Favorite_Genre'].value_counts().index)
plt.title("Top_Genres")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['Device_Used'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Device Distribution")
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Favorite_Genre', y='User_Rating')
plt.xticks(rotation=45)
plt.title("User_Ratings_by_Genre")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Region', hue='Churned')
plt.xticks(rotation=45)
plt.title("Churn_by_Region")
plt.show()
```

```python
#Insight Summary


churn_rate = (df['Churned'] == "Yes").mean() * 100
top_genre = df['Favorite_Genre'].value_counts().idxmax()
top_device = df['Device_Used'].value_counts().idxmax()

print(f" Churn Rate: {churn_rate:.2f}%")
print(f" Most Popular Genre: {top_genre}")
print(f" Most Used Device: {top_device}")
print(f" Avg. User Rating: {df['User_Rating'].mean():.2f}")
print(f" Avg. Watch Hours: {df['Watch_Hours'].mean():.2f}")
```
